```python
###  1. Question 
### Create a Laptop class with a private attribute __price. Implement a method set_price(new_price) to update the price and
## get_price() to retrieve the price. 
```

```python
class Laptop:
    def __init__(self,price):
        self.__price=price
    def set_prie(self,new_price):
        if new_price>price:
            self.__price=new_price
    def get_price(self):
        return self.__price

l1=Laptop("25000")
print(l1.get_price)
```

```python
###  2. Class bank acoount and attribute is deposite and withrawl and print last amount ###
```

```python
class Bankaccount:
    def __init__(self,balance):
        self.__balance=balance
    
    def get_balance(self):
        return self.__balance
    
    def deposite(self,amount):
        if amount>0:
            self.__balance+=amount
            print("Deposite amount is:", amount)
        else:
            print("Invalid deposite")
    def withdrawl(self,amount):
        if amount<=self.__balance:
            self.__balance-=amount
            print('Withdrawl amount is:', amount)
        else:
            print("Invalid Withdrawl")


acc= Bankaccount(1000)
acc.deposite(500)
acc.withdrawl(200)
print("Final balance is:", acc.get_balance())
```

```python
###  Question 3:
### Create a Movie class with a private attribute __rating. Implement methods to set and get the rating,
### ensuring the rating is between 1 and 10.
```

```python
class Movie:
    def __init__(self,rating):
        self.__rating=rating
        
    def set_rating(self,new_rating):
        if 1 <= new_rating >=10:
            self.__rating=new_rating

    def get_rating(self):
        return self.__rating

r=Movie("6")
print(r.get_rating)
            
```
