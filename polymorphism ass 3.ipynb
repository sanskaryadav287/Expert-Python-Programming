```python
class Bird:
    def fly(self):
        print("Bird can fly")

class Penguin(Bird):
    def fly(self):
        print("Penguin can not fly")

b=Bird()
p=Penguin()

b.fly()
p.fly()
```

## 1. Question : Write a python program to demonstrate polymorphism where two different classes have the same method name, and a common functions calls those method. ##

```python
class Cat:
    def sound(self):
        return "Meow"
class Dog:
    def sound(self):
        return "Woof"

cat=Cat()
dog=Dog()

animals=[cat,dog]
for i in animals:
    print(i.sound())
```

### 2. Question : Create a base class Vehicle with a method speed(). Then, create a derived class Car that overrides the speed() method. ##

```python
class Vehicle:
    def speed(self):
        print("The vehicle move at standard speed.")

class Car(Vehicle):
    def speed(self):
        print("The car move at a speed of 120 km/h.")

vehicle=Vehicle()
car=Car()

vehicle.speed()
car.speed()
```
