**Creator an Iterator from a list and print it's first two elements using next()?**

```python
salman = [1,2,3,4,5] 
iterator = iter(salman)
```

```python
print(next(iterator))
print(next(iterator))
```

**Create a string into an iterator and print each character using next()?**

```python
text="Hello"
text_iter=iter(text)
print(next(text_iter))
print(next(text_iter))
```

**Create an iterator and print by using for loop?**

```python
data=[1,2,3,4,5,6]
iterator=iter(data)
for item in iterator:
    print(item)
```

**How to you use a for loop to iterate over an iterator?**

```python
letter=iter(["A","B","C","D"])
for char in letter:
    print(char)
```

**How do you use iter() and next() to iterate through a range of numbers?**
