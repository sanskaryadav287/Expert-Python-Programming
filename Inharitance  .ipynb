```python
## 1.Create an ElectricCar class that inherits from the Car class and includes an additional attribute, battery_size.
```

```python
class Car:
    def __init__(self,model,brand):
        self.model=model
        self.brand=brand

class type(Car):
    def __init__(self,model,brand,battery_size):
        super().__init__(model,brand)
        self.battery_size=battery_size
c1=type("Fortuner","Toyta","80 KW")
print(c1.model,c1.brand,c1.battery_size)
```

```python
### 2. Define a Laptop class that inherits from the Computer class and includes an additional attribute, battery_life.
```

```python
class Laptop:
    def __init__(self,brand,processor):
        self.brand=brand
        self.processor=processor

class Battery(Laptop):
    def __init__(self,brand,processor,battery_life):
        super().__init__(brand,processor)
        self.battery_life=battery_life


l1=Battery("Lenvo","H-Type","6-h")
print(l1.brand,l1.processor,l1.battery_life)
```

```python
## 3. Create a Smartphone class and add another child class of your choice that inherits from an ElectronicDevice class. Include an additional
## attribute, operating_system.
```

```python
class Smartphone:
    def __init__(self,brand,storage):
        self.brand=brand
        self.storage=storage

class device(Smartphone):
    def __init__(self,brand,storage,os):
        super().__init__(brand,storage)
        self.os=os

s1=device("Samsung s25 Ultra","1tb","Andoid 15")
print(s1.brand,s1.storage,s1.os)

```

```python
## 4. Implement a Musician class and create another child class of your choice that extends an Artist class.
```

```python
class Musician:
    def __init__(self,name,age):
        self.name=name
        self.age=age

class Artist(Musician):
    def __init__(self,name,age,play):
        super().__init__(name,age)
        self.play=play

a=Artist("Salman",20,"Tabla")
print(a.name,a.age,a.play)
```

```python
## 5. Define a FictionBook class and add another child class of your choice that inherits from the Book class. Include an additional attribute, genre.
```

```python
class FictionBook:
    def __init__(self,name,writer):
        self.name=name
        self.writer=writer

class Book(FictionBook):
    def __init__(self,name,writer,genre):
        super().__init__(name,writer)
        self.genre=genre

b1=Book("The Jungle book","Akku","Life of Jungle")

print(b1.name,b1.writer,b1.genre)
```

```python
## 6.Create a student class that takes the name & marks of 3 subjects as arguments in the constructor. Then create a method to print the average.

```

```python
class Student:
    def __init__(self,name,m1,m2,m3):
        self.name=name
        self.marks=[m1,2,m3]

s1=Student("Amir",22,25,34)
total=sum(s1.marks)
length=len(s1.marks)
average=total/length
print(s1.name,total,average)

```
